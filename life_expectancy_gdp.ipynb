## Import Modules

First step is to import the necessary Python modules

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

## Load Data

The next step was to create a data frame with the csv file on GDP and life expectancy for each country.

After doing this I wanted to take a look at the first few rows to get a sense of the data and get a list of the countries included.

```python
all_data = pd.read_csv('all_data.csv')
print(all_data.head())
print(all_data.Country.unique())
```

## Observe Data

Before visualizing the data, I wanted to organize it. 

The life expectancy column name was too long, so it was changed to the abbreviation, `LEABY`.

Then I created a data frame for each country.

```python
all_data = all_data.rename({'Life expectancy at birth (years)':'LEABY'}, axis='columns')

chile_df = all_data[all_data.Country == 'Chile']
china_df = all_data[all_data.Country == 'China']
germany_df = all_data[all_data.Country == 'Germany']
mexico_df = all_data[all_data.Country == 'Mexico']
us_df = all_data[all_data.Country == 'United States of America']
zimbabwe_df = all_data[all_data.Country == 'Zimbabwe']
```

The first visualization I wanted to observe was a country by country comparison between their `LEABY` and `GDP` over the documented years. 

```python
country_arr = [chile_df, china_df, germany_df, mexico_df, us_df, zimbabwe_df]
years = range(2000, 2016)

for df in country_arr:
    country = df.Country.values[0]
    
    fig = plt.figure(figsize=(15, 4))
    ax1 = plt.subplot(1,2,1)
    
    x_values1 = df.Year.values
    plt.plot(df.Year, df.LEABY, color='#0C7BDC')
    plt.title(f'{country} LEABY')
    plt.ylabel('Life Expectanvy at Birth (years)')
    plt.xlabel('Year')
    ax1.set_xticks(years)
    plt.xticks(rotation=45)
    
    ax2 = plt.subplot(1,2,2)
    plt.plot(df.Year, df.GDP, color = '#FFC20A')
    plt.title(f'{country} GDP')
    plt.ylabel('GDP')
    plt.xlabel('Year')
    ax2.set_xticks(years)
    plt.xticks(rotation=45)
    
    plt.show()
    plt.clf()
```

These plots show us that as each country's `GDP` increases, their `LEABY` increases as well. It is important to note that the values of the `GDP` and `LEABY` are very different, but they are all positively correlated.

```python

plt.figure(figsize=(8, 6))

sns.lineplot(data=all_data, x='Year', y='LEABY', hue='Country')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

Here we see the `LEABY` for each country over the documented years. Most countries are in the 70's or 80's with Zimbabwe being much lower, but climbing.

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=all_data, x='Year', y='GDP', hue='Country')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

This chart shows the `GDP` for each country. Notice how each countries overall `GDP` is growing overall along with their `LEABY`. For a closer look, check back on the first set of charts that go country by country.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=all_data.LEABY, y=all_data.GDP, hue=all_data.Country)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()

sns.scatterplot(x=zimbabwe_df.LEABY, y=zimbabwe_df.GDP, color='brown')
plt.title('Zimbabwe Scatter')
plt.show()
```

Here we see a scatterplot showing the correlation between `GDP` and `LEABY`. Below the first graph is one for Zimbabwe specifically since it is hard to notice a difference in the first chart. For all countries we see that their `LEABY` increasees as their `GDP` increases overall.
